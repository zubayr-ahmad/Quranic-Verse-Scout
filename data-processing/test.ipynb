```python
import requests
```

```python
url = 'https://api.quran.com/api/v4/search?q=jannat'
```

```python
response = requests.get(url)
```

```python
data = response.json()
data
```

```python
import pandas as pd
df = pd.DataFrame(data['search']['results'][0]['translations'])
df.head()
```

```python
# function remove <> and everthing in between
def remove_html_tags(text):
    import re
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
# remove from text
df['text_clean'] = df['text'].apply(remove_html_tags)
df.head()
```
